### Converting catergorical into numerical

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
le = LabelEncoder()
df1['Sex'] = le.fit_transform(df['Sex'])
```

```python
df1
```

```python
df1['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df1
```

```python
from sklearn.preprocessing import MinMaxScaler
numeric_cols = df1.select_dtypes(include=['float64', 'int64']).columns

scaler = MinMaxScaler()
df1[numeric_cols] = scaler.fit_transform(df1[numeric_cols])
```

```python
df1
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for each numeric column
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df1[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
# Removing outliers

for col in numeric_cols:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out the outliers
    df1 = df1[(df1[col] >= lower_bound) & (df1[col] <= upper_bound)]
```

```python
# Boxplot for each numeric column
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df1[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```
